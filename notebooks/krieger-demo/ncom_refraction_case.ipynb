```python
import numpy as np
import xarray as xr
import matplotlib.pyplot as plt
import pandas as pd

import sys
sys.path.append('/Users/jkrieger/work/ray_tracing/python')
import mantaray
```

```python
g = 9.8
def period2wavenumber(T):
    k = (2*np.pi)**2/g/T**2
    return k

def group_velocity(k):
    cg = 0.5*(g/k)**.5
    return cg
```

## Define Currents

```python
dataset = xr.load_dataset('data/currents/ncom_refraction_slice.nc')
#smaller_spatial_slice = dataset.sel(lon=slice(234.5, 237), lat=slice(36.5,38.5))

# Extract lat/lon
longitudes = dataset.lon
latitudes = dataset.lat

# Convert longitudes and latitudes to meters
Lon_meter = ((longitudes - np.mean(longitudes)) * 1_852 * 60 * np.cos(np.mean(latitudes) * np.pi / 180)) / 1_000
Lat_meter = ((latitudes - np.mean(latitudes)) * 1_852 * 60) / 1_000

# Shift so the max value is at 0
Lon_meter += Lon_meter.max()
Lat_meter += Lat_meter.max()

x = Lon_meter.values
y = Lat_meter.values

# Extract u and v components from the dataset
u_component = dataset["water_u"].isel(time=0).values
v_component = dataset["water_v"].isel(time=0).values

currents = np.sqrt(u_component**2 + v_component**2)
Umax = currents.max()

# Create an xarray dataset following the reference format
ds = xr.Dataset(
    data_vars={
        "x": (["x"], x),
        "y": (["y"], y),
        "u": (["y", "x"], u_component, {"long_name": "u", "units": "m/s"}),
        "v": (["y", "x"], v_component, {"long_name": "v", "units": "m/s"}),
    })

# Save as NetCDF
#ds.to_netcdf("output_ncom_refraction_current.nc", format="NETCDF3_CLASSIC")
```

## Define Bathymetry

```python
# Number of points in x and y
nx, ny = len(x), len(y)

xc = np.linspace(0, 1, ny)
u = Umax*xc
u = np.tile(u, (nx, 1)).T
v = np.zeros((ny, nx))

left_quarter_index = nx // 4
u[:, :left_quarter_index] = 0

output_bathy = "data/bathymetry/bathy_constant.nc"

# Considering a constant depth of 4km
D = 4e3 
depth = D*np.ones((ny, nx))

ds = xr.Dataset(
    data_vars={
        "x": (["x"], x),
        "y": (["y"], y),
        "depth": (["y", "x"], depth, {"long_name": "depth", "units": "m"}),
    })

#ds.to_netcdf(output_bathy, format="NETCDF3_CLASSIC")
```

```python
# Period of incident waves in seconds
T0 = 11
# Direction of incident waves in radians (trig convention, going to)
theta0 = (5.3*(180/np.pi))-np.pi
# Convert period to wavenumber magnitude
k0 = period2wavenumber(T0)
# Calculate wavenumber components
kx0 = k0*np.cos(theta0)
ky0 = k0*np.sin(theta0)

# Number of rays
n_rays = 100

# Initialize wavenumber for all rays
Kx0 = kx0*np.ones(n_rays)
Ky0 = ky0*np.ones(n_rays)

# Current and bathymetry file path
bathymetry = 'data/bathymetry/bathy_zonal_jet.nc'
ncom = 'data/currents/output_ncom_refraction_current_large_scale.nc'

ds = xr.open_dataset(ncom)
x = ds.x.values
y = ds.y.values

# Creates initial x position for all rays
x0 = 10*np.ones(n_rays)
y0 = np.linspace(0, y.max(), n_rays)

# Estimates CFL
# Computes grid smallest spacing
dd = np.min([np.diff(x).mean(), np.diff(y).mean()])
# Computes group velocity
cg = group_velocity(k0)
# Computes CFL
cfl = dd/cg

duration = round(x.max()/cg)
step_size = cfl
```

```python
x0 = np.linspace(0, 400, num=100)  # 100 rays spread from 0 to 125 in x
y0 = np.full_like(x0, 1200)  # Set y positions at 220
bundle = mantaray.ray_tracing(x0, y0, Kx0, Ky0, duration, step_size, bathymetry, ncom)
```
